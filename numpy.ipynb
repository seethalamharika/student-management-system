```python
import numpy as np
marks=np.array([56,76,98,65,45])
print("Average:",np.mean(marks))
print("highest:",np.max(marks))
print("lowest:",np.min(marks))
top_students=marks[marks>75]
print(top_students)
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,3,4,5,])
y=np.array([10,20,30,40,50])
plt.plot(x,y)
plt.title("Line chart")
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,3,4,5,])
y=np.array([10,20,30,40,50])
plt.bar(x,y)
plt.title("Line chart")
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```
